N-Grams using pattern 
basically we are making tri-grams here 


```python
from pattern.en import ngrams

print(ngrams("There is nothing either good or bad, but thinking makes it so.", n=3))
```

Bi-Grams using Pattern

```python
print(ngrams("There is nothing either good or bad, but thinking makes it so.", n=2))
```

Quantify : This function gives a word count estimation of the words given

```python
from pattern.en import quantify
a = quantify(['Car', 'bike', 'Bus', 'Truck', 'Train', 'Car', 'tractor', 'Bus', 'Truck'])
print(a)
```

```python
print(quantify({'Bus': 100, 'Car': 1500,'Truck': 700}))
print(quantify('People', amount=60000))
```

```python
from pattern.en import number, numerals
```

```python
print(number("ten thousand five hundred seventy one"))
```

```python
print(numerals(10571))
print(numerals(567.79585, round=3))
```

```python
from pattern.en import pluralize,singularize
print(pluralize('boy'))
print(singularize('men'))
```

```python
print(pluralize('Student'))
print(singularize('chocolates'))
```

```python
from pattern.en import comparative, superlative
print(comparative('hungry'))
print(superlative('hungry'))
```

Data Mining Using Pattern library 

```python
from pattern.web import Google
google = Google()

for result in google.search('Tvs'):
    print(result.url)
    print(result.text)
```

```python
for result in google.search('NASA'):
    print(result.url)
    print(result.text)
```
