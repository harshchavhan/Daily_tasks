```python
def cleaner(instance):
    name = instance
    name = str(name).lower()
    # name = re.sub(r'\s*$', ' ', name)
    name = re.sub(r' ', ' ', name) #Remove multiple spaces with single space
    name = re.sub(r'^s/o ', '', name) # remove starting 's/o'
    name = re.sub(r'^mr', '', name) # remove starting 'mr'
    name = re.sub(r'[0-9]', '', name) #Remove Numbers
    name = re.sub(r'[/]', '', name) #Remove Numbers
    # name = re.sub(r'@[a-zA-Z0-9-]+\.[a-zA-Z0-9-._]+', '', name) #Convert email to name
    name = " ".join(name.split(".")) # Remove dot from the instances
    name = ' '.join( [w for w in name.split() if len(w)>2] )# Remove single and double characters.

    return name
```
